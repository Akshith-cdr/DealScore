## Metric Meaning

- Accuracy: Overall percentage of correct labels.
- Precision: Of the listings predicted as a category, how many were correct.
- Recall: Of the listings that truly belong to a category, how many were found.
- Confusion Matrix: A table showing where the classifier is correct or confused.
