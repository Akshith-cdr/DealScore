# 01 - Exploratory Data Analysis

This notebook explores the cleaned Indian used-car dataset before model training.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

ROOT_DIR = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
DATA_PATH = ROOT_DIR / 'data' / 'processed' / 'cars_cleaned.csv'

sns.set_theme(style='whitegrid')
df = pd.read_csv(DATA_PATH)
df.head()
```

## Dataset Shape and Missing Values

This confirms how many rows and columns are available after preprocessing, and whether missing values remain.

```python
print(df.shape)
df.isna().sum()
```

## Price Distribution

This graph shows how used-car prices are spread. A right-skewed distribution is common because most cars are affordable, while a few premium cars are very expensive.

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['price'], bins=40, kde=True)
plt.title('Price Distribution')
plt.xlabel('Price (INR)')
plt.ylabel('Number of Cars')
plt.show()
```

## Brand Analysis

This chart shows which brands appear most often in the dataset. Popular brands usually give the model more examples to learn from.

```python
top_brands = df['brand'].value_counts().head(15)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_brands.values, y=top_brands.index)
plt.title('Top 15 Brands by Listing Count')
plt.xlabel('Number of Listings')
plt.ylabel('Brand')
plt.show()
```

## Correlation Heatmap

The heatmap shows relationships between numeric features. Strong positive or negative correlations can help explain model behavior.

```python
numeric_df = df.select_dtypes(include='number')
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Numeric Feature Correlation Heatmap')
plt.show()
```

## KM Driven vs Price

This scatter plot checks whether heavily driven cars tend to have lower prices.

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x='km_driven', y='price', alpha=0.5)
plt.title('KM Driven vs Price')
plt.xlabel('KM Driven')
plt.ylabel('Price (INR)')
plt.show()
```

## Fuel Type Analysis

This boxplot compares price ranges across fuel types. It helps identify whether petrol, diesel, CNG, or electric cars differ in market value.

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='fuel_type', y='price')
plt.title('Fuel Type vs Price')
plt.xlabel('Fuel Type')
plt.ylabel('Price (INR)')
plt.xticks(rotation=30)
plt.show()
```
